```python
q1=np.percentile(df["Dispatch Qty"],25,method='midpoint')
q3=np.percentile(df["Dispatch Qty"],75,method='midpoint')
```

```python
iqr=q3-q1
```

```python
iqr
```

```python
upper=q3+1.5*iqr
lower=q1-1.5*iqr
```

```python
upper_arr=np.where(df["Dispatch Qty"]>=upper)[0]
lower_arr=np.where(df["Dispatch Qty"]<=lower)[0]
```

```python
df.drop(index=upper_arr,inplace=True)
df.drop(index=lower_arr,inplace=True)
```
